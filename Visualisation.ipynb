```python
import pandas as pd
# Load EV dataset
data = pd.read_csv('Electric_Vehicle_Population_Cleaned.csv')
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['Electric Vehicle Type'] = le.fit_transform(data['Electric Vehicle Type'])
```

```python
# Remove duplicate entries
data.drop_duplicates(inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data[['Electric Range', 'Base MSRP']] = scaler.fit_transform(data[['Electric Range', 'Base MSRP']])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.countplot(x='Electric Vehicle Type', data=data)
plt.title('Distribution of EV Types')
plt.xlabel('Vehicle Type')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(data['Electric Range'], kde=True, bins=20, color='blue')
plt.title('Distribution of Electric Vehicle Ranges')
plt.xlabel('Range (miles)')
plt.ylabel('Frequency')
plt.show()
```
